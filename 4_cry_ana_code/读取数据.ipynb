```python
file_path = r'C:/Users/zhang/Desktop/md/2024corning/july/tension_las/14stress_y.txt'
```

```python
def read_data(file_path):
    data = []
    with open(file_path, 'r') as file:
        for line in file:
            line = line.strip()
            # 尝试将字符串转换为float类型
            try:
                value = float(line)
            except ValueError:
                value = line  # 如果转换失败，则保持为字符串类型
            data.append(value)
    return data

# 使用示例
data_list = read_data(file_path)
count = 0
for i in data_list:
    print(i)
    count+=1
    
```
