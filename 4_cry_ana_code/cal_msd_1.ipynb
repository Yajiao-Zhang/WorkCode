```python
def read_xyz_frames(file_path):
    frames = []
    current_frame = None

    with open(file_path, 'r') as file:
        lines = file.readlines()
        line_index = 0

        while line_index < len(lines):
            line = lines[line_index].strip()

            if line.isdigit():
                # 读取原子数
                if current_frame:
                    frames.append(current_frame)
                num_atoms = int(line)
                current_frame = {"num_atoms": num_atoms, "atoms": []}
                line_index += 2  # 跳过标题行 "Atoms"
            else:
                # 读取原子坐标
                if current_frame:
                    atom_data = line.split()
                    if len(atom_data) == 4:
                        element, x, y, z = atom_data
                        current_frame["atoms"].append({
                            "element": element,
                            "x": float(x),
                            "y": float(y),
                            "z": float(z)
                        })
                line_index += 1

        if current_frame:
            frames.append(current_frame)

    return frames

# 使用之前的frames列表
frames = read_xyz_frames(file_path)

# 找出最大的x、y和z值
max_x = max([max(frame["atoms"], key=lambda atom: atom["x"])["x"] for frame in frames])
max_y = max([max(frame["atoms"], key=lambda atom: atom["y"])["y"] for frame in frames])
max_z = max([max(frame["atoms"], key=lambda atom: atom["z"])["z"] for frame in frames])

print("Maximum X value:", max_x)
print("Maximum Y value:", max_y)
print("Maximum Z value:", max_z)
```

```python
 # 模拟中的周期性边界条件，假设每个维度的边界大小
box_size = np.array([113.7, 24.0, 74.8])  # 根据您的模拟设置
```

```python
def calculate_msd(frames, element, box_sizes):
    msd = np.zeros(len(frames))  # 存储每个时间步的MSD
    ref_frame = frames[0]  # 参考帧，t=0

    for t, frame in enumerate(frames):
        squared_displacements = []

        for atom_t, atom_ref in zip(frame["atoms"], ref_frame["atoms"]):
            # 考虑周期性边界条件，分别对X、Y、Z轴进行处理
            dx = atom_t["x"] - atom_ref["x"]
            dy = atom_t["y"] - atom_ref["y"]
            dz = atom_t["z"] - atom_ref["z"]

            # 分别对X、Y、Z轴应用周期性边界条件
            for i in range(3):
                cell_size = box_sizes[i]  # 每个轴上的晶格大小
                delta = [dx, dy, dz][i]
                delta -= cell_size * round(delta / cell_size)

                # 重新赋值回 dx, dy, dz
                if i == 0:
                    dx = delta
                elif i == 1:
                    dy = delta
                elif i == 2:
                    dz = delta

            squared_displacement = dx ** 2 + dy ** 2 + dz ** 2

            if atom_t["element"] == element:
                squared_displacements.append(squared_displacement)

        if squared_displacements:
            msd[t] = np.mean(squared_displacements)
        else:
            msd[t] = 0.0

    return msd

# 使用之前的frames列表
box_sizes = box_size # 根据你的数据提供晶格大小
o_msd = calculate_msd(frames, element="O", box_sizes=box_sizes)
nb_msd = calculate_msd(frames, element="Nb", box_sizes=box_sizes)
li_msd = calculate_msd(frames, element="Li", box_sizes=box_sizes)
```

```python
import numpy as np

def calculate_element_rmsd(reference_frame, frames, box_sizes, element):
    rmsd_values = []

    for frame in frames:
        num_atoms = frame["num_atoms"]
        reference_atoms = [atom for atom in reference_frame["atoms"] if atom["element"] == element]
        current_atoms = [atom for atom in frame["atoms"] if atom["element"] == element]

        if len(reference_atoms) != len(current_atoms):
            rmsd_values.append(0.0)
            continue

        squared_distances = []

        for i in range(len(reference_atoms)):
            dx = reference_atoms[i]["x"] - current_atoms[i]["x"]
            dy = reference_atoms[i]["y"] - current_atoms[i]["y"]
            dz = reference_atoms[i]["z"] - current_atoms[i]["z"]

            # 考虑周期性边界条件
            for j in range(3):
                cell_size = box_sizes[j]
                delta = [dx, dy, dz][j]
                delta -= cell_size * round(delta / cell_size)
                [dx, dy, dz][j] = delta

            squared_distance = dx ** 2 + dy ** 2 + dz ** 2
            squared_distances.append(squared_distance)

        rmsd = np.sqrt(np.mean(squared_distances))
        rmsd_values.append(rmsd)

    return rmsd_values

# 使用之前的frames列表
box_sizes = box_size  # 根据你的数据提供晶格大小
reference_frame = frames[0]  # 选择一个帧作为参考帧

o_rmsd = calculate_element_rmsd(reference_frame, frames, box_sizes, element="O")
nb_rmsd = calculate_element_rmsd(reference_frame, frames, box_sizes, element="Nb")
li_rmsd = calculate_element_rmsd(reference_frame, frames, box_sizes, element="Li")

# 打印三者之间的RMSD值
```

```python
def calculate_rmsd_between_frames(frame1, frame2, box_sizes):
    num_atoms = frame1["num_atoms"]
    atoms1 = frame1["atoms"]
    atoms2 = frame2["atoms"]

    squared_distances = []

    for i in range(num_atoms):
        dx = atoms1[i]["x"] - atoms2[i]["x"]
        dy = atoms1[i]["y"] - atoms2[i]["y"]
        dz = atoms1[i]["z"] - atoms2[i]["z"]

        # 考虑周期性边界条件
        for j in range(3):
            cell_size = box_sizes[j]
            delta = [dx, dy, dz][j]
            delta -= cell_size * round(delta / cell_size)
            [dx, dy, dz][j] = delta

        squared_distance = dx ** 2 + dy ** 2 + dz ** 2
        squared_distances.append(squared_distance)

    rmsd = np.sqrt(np.mean(squared_distances))
    return rmsd

def calculate_rmsd_for_frames(frames, box_sizes):
    rmsd_values = [0.0]  # 初始帧的RMSD为0

    for i in range(1, len(frames)):
        rmsd = calculate_rmsd_between_frames(frames[i - 1], frames[i], box_sizes)
        rmsd_values.append(rmsd)

    return rmsd_values

# 使用之前的frames列表
box_sizes = box_size  # 根据你的数据提供晶格大小
rmsd_values = calculate_rmsd_for_frames(frames, box_sizes)

# 绘制RMSD图
time_steps = list(range(len(rmsd_values)))
plt.plot(time_steps, rmsd_values, label='RMSD', color='blue')
plt.xlabel('时间步骤')
plt.ylabel('RMSD')
plt.title('每帧与前一帧的RMSD')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_rmsd_between_frames(frame1, frame2, box_sizes, element):
    num_atoms = frame1["num_atoms"]
    atoms1 = [atom for atom in frame1["atoms"] if atom["element"] == element]
    atoms2 = [atom for atom in frame2["atoms"] if atom["element"] == element]

    if len(atoms1) != len(atoms2):
        return 0.0  # 如果元素数量不匹配，则RMSD为0

    squared_distances = []

    for i in range(len(atoms1)):
        dx = atoms1[i]["x"] - atoms2[i]["x"]
        dy = atoms1[i]["y"] - atoms2[i]["y"]
        dz = atoms1[i]["z"] - atoms2[i]["z"]

        # 考虑周期性边界条件
        for j in range(3):
            cell_size = box_sizes[j]
            delta = [dx, dy, dz][j]
            delta -= cell_size * round(delta / cell_size)
            [dx, dy, dz][j] = delta

        squared_distance = dx ** 2 + dy ** 2 + dz ** 2
        squared_distances.append(squared_distance)

    rmsd = np.sqrt(np.mean(squared_distances))
    return rmsd

def calculate_element_rmsd_diff(frames, box_sizes, element):
    rmsd_values = [0.0]  # 初始帧的RMSD为0

    for i in range(1, len(frames)):
        rmsd = calculate_rmsd_between_frames(frames[i - 1], frames[i], box_sizes, element)
        rmsd_values.append(rmsd)

    return rmsd_values

# 使用之前的frames列表
box_sizes = box_size  # 根据你的数据提供晶格大小

# 计算不同元素与上一帧的RMSD差异
element_names = ["O", "Nb", "Li"]
element_rmsd_diff = {}
for element in element_names:
    element_rmsd_diff[element] = calculate_element_rmsd_diff(frames, box_sizes, element)

# 绘制不同元素的RMSD差异图
time_steps = list(range(len(frames)))
for element in element_names:
    plt.plot(time_steps, element_rmsd_diff[element], label=f'{element} RMSD', alpha=0.7)

plt.xlabel('时间步骤')
plt.ylabel('RMSD差异')
plt.title('不同元素与上一帧的RMSD差异')
plt.legend()
plt.grid(True)
plt.show()
```
