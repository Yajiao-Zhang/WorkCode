```python
import numpy as np

def calculate_atom_spacing(num_atoms, box_size):
    """
    计算给定原子数和盒子边长的每个原子间距。
    
    参数:
    num_atoms (int): 原子总数
    box_size (float): 盒子边长
    
    返回:
    float: 每个原子的间距
    """
    # 计算每个维度的原子数量
    grid_size = int(np.ceil(num_atoms ** (1/3)))  # 计算每个维度的网格数量

    # 计算每个原子的间距
    spacing = box_size / grid_size
    return spacing

```

```python
# 示例调用
num_atoms = 30000
box_size = 75.22
spacing = calculate_atom_spacing(num_atoms, box_size)
print(f"每个原子的间距: {spacing:.3f} 单位")
```
