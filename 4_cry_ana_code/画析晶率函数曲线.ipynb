```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_min = 1100
x_max = 2100
```

```python
def fun_viscosity(t):
    return np.exp(-5.12+9829.41/(t-378.94))
```

```python
def fun_temp(t,Tm):
    return 1-t/Tm
```

```python
def fun_all(t,Tm):
    return (1/fun_viscosity(t))*fun_temp(t,Tm)
```

```python
x_list = np.linspace(x_min,x_max,1001)
```

```python
y_list = fun_all(x_list,2100)
```

```python
max(y_list)
```

```python
y_list = (y_list-min(y_list))/max(y_list)
```

```python
x1,y1 = [],[]
x_origin =np.array([580,595,648,887,1171,1633,2262,2244,1599,799,580])
xx = 1200
for i in x_origin:
    x1.append(xx)
    y1.append(round((i-580)/(2262-580),4))
    print(xx,round((i-580)/(2262-580),4))
    xx+=100
```
