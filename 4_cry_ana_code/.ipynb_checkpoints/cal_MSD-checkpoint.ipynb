```python
def read_xyz_frames(file_path):
    frames = []
    current_frame = None

    with open(file_path, 'r') as file:
        lines = file.readlines()
        line_index = 0

        while line_index < len(lines):
            line = lines[line_index].strip()

            if line.isdigit():
                # 读取原子数
                if current_frame:
                    frames.append(current_frame)
                num_atoms = int(line)
                current_frame = {"num_atoms": num_atoms, "atoms": []}
                line_index += 2  # 跳过标题行 "Atoms"
            else:
                # 读取原子坐标
                if current_frame:
                    atom_data = line.split()
                    if len(atom_data) == 4:
                        element, x, y, z = atom_data
                        current_frame["atoms"].append({
                            "element": element,
                            "x": float(x),
                            "y": float(y),
                            "z": float(z)
                        })
                line_index += 1

        if current_frame:
            frames.append(current_frame)

    return frames

# 使用之前的frames列表
frames = read_xyz_frames(file_path)

# 找出最大的x、y和z值
max_x = max([max(frame["atoms"], key=lambda atom: atom["x"])["x"] for frame in frames])
max_y = max([max(frame["atoms"], key=lambda atom: atom["y"])["y"] for frame in frames])
max_z = max([max(frame["atoms"], key=lambda atom: atom["z"])["z"] for frame in frames])

print("Maximum X value:", max_x)
print("Maximum Y value:", max_y)
print("Maximum Z value:", max_z)
```

```python
import numpy as np

def calculate_msd(frames, element, box_sizes):
    msd = np.zeros(len(frames))  # 存储每个时间步的MSD
    ref_frame = frames[0]  # 参考帧，t=0

    for t, frame in enumerate(frames):
        squared_displacements = []
        if t%100==0: #and t>=100 :
            print(t)

        for atom_t, atom_ref in zip(frame["atoms"], ref_frame["atoms"]):
            # 考虑周期性边界条件，分别对X、Y、Z轴进行处理
            dx = atom_t["x"] - atom_ref["x"]
            dy = atom_t["y"] - atom_ref["y"]
            dz = atom_t["z"] - atom_ref["z"]

            # 分别对X、Y、Z轴应用周期性边界条件
            for i in range(3):
                cell_size = box_sizes[i]  # 每个轴上的晶格大小
                delta = [dx, dy, dz][i]
                delta -= cell_size * np.round(delta / cell_size)

                # 重新赋值回 dx, dy, dz
                if i == 0:
                    dx = delta
                elif i == 1:
                    dy = delta
                elif i == 2:
                    dz = delta

            squared_displacement = dx ** 2 + dy ** 2 + dz ** 2

            if atom_t["element"] == element:
                squared_displacements.append(squared_displacement)

        if squared_displacements:
            msd[t] = np.mean(squared_displacements)
        else:
            msd[t] = 0.0

    return msd

# 使用之前的frames列表
box_sizes = box_size # 根据你的数据提供晶格大小
o_msd = calculate_msd(frames, element="O", box_sizes=box_sizes)
si_msd = calculate_msd(frames, element="Si", box_sizes=box_sizes)
al_msd = calculate_msd(frames, element="Al", box_sizes=box_sizes)
li_msd = calculate_msd(frames, element="Li", box_sizes=box_sizes)
```

```python
import numpy as np

def calculate_element_rmsd(reference_frame, frames, box_sizes, element):
    rmsd_values = []

    for frame in frames:
        num_atoms = frame["num_atoms"]
        reference_atoms = [atom for atom in reference_frame["atoms"] if atom["element"] == element]
        current_atoms = [atom for atom in frame["atoms"] if atom["element"] == element]

        if len(reference_atoms) != len(current_atoms):
            rmsd_values.append(0.0)
            continue

        squared_distances = []

        for i in range(len(reference_atoms)):
            dx = reference_atoms[i]["x"] - current_atoms[i]["x"]
            dy = reference_atoms[i]["y"] - current_atoms[i]["y"]
            dz = reference_atoms[i]["z"] - current_atoms[i]["z"]

            # 考虑周期性边界条件
            for j in range(3):
                cell_size = box_sizes[j]
                delta = [dx, dy, dz][j]
                delta -= cell_size * round(delta / cell_size)
                [dx, dy, dz][j] = delta

            squared_distance = dx ** 2 + dy ** 2 + dz ** 2
            squared_distances.append(squared_distance)

        rmsd = np.sqrt(np.mean(squared_distances))
        rmsd_values.append(rmsd)

    return rmsd_values

# 使用之前的frames列表
box_sizes = box_size  # 根据你的数据提供晶格大小
reference_frame = frames[0]  # 选择一个帧作为参考帧

o_rmsd = calculate_element_rmsd(reference_frame, frames, box_sizes, element="O")
nb_rmsd = calculate_element_rmsd(reference_frame, frames, box_sizes, element="Nb")
li_rmsd = calculate_element_rmsd(reference_frame, frames, box_sizes, element="Li")

# 打印三者之间的RMSD值
```

```python
def calculate_rmsd_between_frames(frame1, frame2, box_sizes):
    num_atoms = frame1["num_atoms"]
    atoms1 = frame1["atoms"]
    atoms2 = frame2["atoms"]

    squared_distances = []

    for i in range(num_atoms):
        dx = atoms1[i]["x"] - atoms2[i]["x"]
        dy = atoms1[i]["y"] - atoms2[i]["y"]
        dz = atoms1[i]["z"] - atoms2[i]["z"]

        # 考虑周期性边界条件
        for j in range(3):
            cell_size = box_sizes[j]
            delta = [dx, dy, dz][j]
            delta -= cell_size * round(delta / cell_size)
            [dx, dy, dz][j] = delta

        squared_distance = dx ** 2 + dy ** 2 + dz ** 2
        squared_distances.append(squared_distance)

    rmsd = np.sqrt(np.mean(squared_distances))
    return rmsd

def calculate_rmsd_for_frames(frames, box_sizes):
    rmsd_values = [0.0]  # 初始帧的RMSD为0

    for i in range(1, len(frames)):
        rmsd = calculate_rmsd_between_frames(frames[i - 1], frames[i], box_sizes)
        rmsd_values.append(rmsd)

    return rmsd_values

# 使用之前的frames列表
box_sizes = box_size  # 根据你的数据提供晶格大小
rmsd_values = calculate_rmsd_for_frames(frames, box_sizes)

# 绘制RMSD图
time_steps = np.linspace(0,30,500)
plt.plot(time_steps, rmsd_values, label='RMSD', color='blue')
plt.xlabel('时间步骤')
plt.ylabel('RMSD')
plt.title('每帧与前一帧的RMSD')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

def calculate_distance(coord1, coord2, box_size):
    """计算两个原子之间的距离，考虑周期性边界条件"""
    delta = np.abs(coord1 - coord2)
    delta = np.where(delta > 0.5 * box_size, box_size - delta, delta)
    distance = np.sqrt(np.sum(delta**2))
    return distance

def find_neighbors(atom_list, box_size, cutoff_distance):
    """找到每个原子的近邻"""
    num_atoms = len(atom_list)
    neighbors_table = {i: [] for i in range(num_atoms)}
    box_size = np.array(box_size)

    for i in range(num_atoms):
        for j in range(i + 1, num_atoms):
            coord_i = np.array(atom_list[i][1:])
            coord_j = np.array(atom_list[j][1:])

            # 计算考虑周期性边界条件的距离
            distance = calculate_distance(coord_i, coord_j, box_size)

            # 如果距离小于阈值，则将 j 添加到 i 的近邻列表中
            if distance < cutoff_distance:
                neighbors_table[i].append(j)
                neighbors_table[j].append(i)

    return neighbors_table

# 示例使用
atom_list = atom_list

box_size = box_size_new  # 替换为你的盒子尺寸
cutoff_distance = 3.0  # 替换为你的近邻阈值

neighbors_table = find_neighbors(atom_list, box_size, cutoff_distance)
print(neighbors_table)
```

```python
def calculate_total_average_coordination_number(neighbors_table):
    """计算总体的平均配位数"""
    total_coordination = sum(len(neighbors) for neighbors in neighbors_table.values())
    total_atoms = len(neighbors_table)
    total_average_coordination = total_coordination / total_atoms
    return total_average_coordination

# 计算总体的平均配位数
total_average_coordination = calculate_total_average_coordination_number(neighbors_table)

print("总体的平均配位数:", total_average_coordination)
```

```python
import numpy as np
import networkx as nx
from collections import Counter


class CycleCalculator:
    def __init__(self):
        pass

    def preprocess(self, adjacency_list):
        G = nx.Graph(adjacency_list)
        adj_matrix = nx.to_numpy_matrix(G)
        np.fill_diagonal(adj_matrix, 0)
        # adj_matrix = np.triu(adj_matrix)
        source_mapping = {node:index for index, node in enumerate(G.nodes())}
        dest_mapping = {index:node for index, node in enumerate(G.nodes())}
        return adj_matrix, source_mapping, dest_mapping
    
    def get_cycles(self, adj_matrix):
        G = nx.from_numpy_matrix(adj_matrix, create_using=nx.DiGraph)
        cycles = list(nx.simple_cycles(G))
        cycles = [cycle for cycle in cycles if len(cycle) > 2]
        return G, cycles

    def calculate_graph(self, adjacency_list, return_summary = False):
        adj_matrix, source_mapping, dest_mapping =  self.preprocess(adjacency_list)
        G, cycles = self.get_cycles(adj_matrix)
        shortest_lengthes = nx.floyd_warshall_numpy(G)
        assert np.all(shortest_lengthes == shortest_lengthes.T), "The shortest length matrix is not symmetric, your adjacency_list is not complete!"
        shortest_lengthes_single_cycles = self.get_cycle_shortest_length(cycles, len(adjacency_list))
        valid_cycles = self.get_valid_cycles(cycles, shortest_lengthes_single_cycles, shortest_lengthes)
        cycles = [cycle for cycle, valid_cycle in zip(cycles, valid_cycles) if valid_cycle]
        cycles = self.get_origin_cycles(cycles, source_mapping, dest_mapping)
        cycles = [cycle for cycle, item in zip(cycles, self.get_valid_not_bigraph(cycles)) if item]
        if return_summary:
            return dict(Counter([len(cycle) for cycle in cycles]))
        else:
            return cycles

    def get_valid_not_bigraph(self, cycles):
        results = []
        cycles = [sorted(cycle) for cycle in cycles]
        cycles_set = set()
        for cycle in cycles:
            if not tuple(cycle) in cycles_set:
                results.append(True)
                cycles_set.add(tuple(cycle))
            else:
                results.append(False)
        return results

    def get_origin_cycles(self, cycles, source_mapping, dest_mapping):
        results = []
        for cycle in cycles:
            results.append([dest_mapping[node] for node in cycle])
        return results
    
    def get_valid_cycles(self, cycles, shortest_lengthes_single_cycles, shortest_lengthes):
        results = []
        for cycle, shortest_lengthes_single_cycle in zip(cycles, shortest_lengthes_single_cycles):
            grid_x, grid_y = np.meshgrid(cycle, cycle, indexing='ij')
            cycle_length_matrix = shortest_lengthes_single_cycle[grid_x, grid_y]
            graph_length_matrix = shortest_lengthes[grid_x, grid_y]
            results.append(True) if np.all(cycle_length_matrix == graph_length_matrix) else results.append(False)
        return results


    def get_cycle_shortest_length(self, cycles, node_length):
        results = []
        for cycle in cycles:
            cycle = np.array(cycle)
            n = len(cycle)
            indexes = np.arange(n)
            result = np.ones((node_length, node_length)) * np.inf
            for i in range(n):
                for j in range(i, n):
                    result[cycle[i], cycle[j]] = result[cycle[j], cycle[i]] = min(abs(indexes[i] - indexes[j]), n - abs(indexes[i] - indexes[j]))
            results.append(result)
        return results


if __name__ == "__main__":
    cycle_calculator = CycleCalculator()
    adjacency_list = neighbors_table
    cycle_list = cycle_calculator.calculate_graph(adjacency_list, return_summary=False)
    cycle_summary = cycle_calculator.calculate_graph(adjacency_list, return_summary=True)
    print(cycle_list, cycle_summary)
```
