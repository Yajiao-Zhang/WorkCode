```python
import numpy as np
```

```python
file_path = r'C:\Users\zhang\Desktop\md\linbo3\26k_nvt_out\log.lammps'
save_path = r'C:\Users\zhang\Desktop\md\linbo3\26k_nvt_out\data_log.txt'
```

```python
list_select = np.linspace(870000,8870000,201)
```

```python
list_select
```

```python
# 打开原始文本文件
with open(file_path, 'r', encoding='utf-8') as input_file:
    # 创建一个新的文件用于存储符合条件的行
    with open(save_path, 'w') as output_file:
        for line in input_file:
             # 忽略空行
            if not line.strip():
                continue
            # 解析每行的首个数字
            first_char = line.split()[0]
            # 检查首个数字是否在指定范围内
            if first_char.isdigit() and int(first_char) in list_select:
            # 如果在范围内，将该行写入log.txt文件
                output_file.write(line)

# 关闭文件
input_file.close()
output_file.close()
```
