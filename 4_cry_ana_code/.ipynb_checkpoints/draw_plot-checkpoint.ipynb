```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 首先是看均值和峰值有没有关系
mean_15k =[1042,1068,1078,1064,1084,1058,1061,1077,1066,1072,1095] 
max_15k = [61,97,101,79,109,82,74,110,73,82,99]
mean_15k = np.array(mean_15k)
max_15k = np.array(max_15k)
```

```python
plt.plot(range(len(mean_15k)),mean_15k-1000,label = '15-mean')
plt.plot( range(len(max_15k)),max_15k, label = '15-max')
plt.legend()
```

```python
mean_16k = [1047,1055,1065,1061,1061,1069,1079,1065,1080,1061,1060]
max_16k = [78,79,76,85,79,84,102,79,90,76,80]
mean_16k = np.array(mean_16k)
max_16k = np.array(max_16k)
plt.plot(range(len(mean_16k)),mean_16k-1000,label = '16-mean')
plt.plot( range(len(max_16k)),max_16k, label = '16-max')
plt.legend()
```
