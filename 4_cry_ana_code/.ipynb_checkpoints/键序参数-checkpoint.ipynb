# 计算Q的公式

```python
# 计算球谐函数 Ylm
def spherical_harmonic(l, m, theta, phi):
    prefactor = np.sqrt((2 * l + 1) * np.math.factorial(l - abs(m)) / (4 * np.pi * np.math.factorial(l + abs(m))))
    legendre_term = lpmv(abs(m), l, np.cos(theta))
    exponential_term = np.exp(1j * m * phi)
    return prefactor * legendre_term * exponential_term
```

```python
# 计算每个原子的qlm
def calculate_qlm(center_pos, neibor_pos, l, m):
    center_pos,neibor_pos = np.array(center_pos), np.array(neibor_pos)
    nb = len(neibor_pos)
    qlm_values = 0.0
    for nei_atom in neibor_pos:
        rij = nei_atom - center_pos    
        theta = np.arccos(rij[2] / np.linalg.norm(rij))
        phi = np.arctan2(rij[1], rij[0])
        qlm_values += spherical_harmonic(l, m, theta, phi)
    return qlm_values / nb
```

```python
# 每个Qlm是四周所有原子K的Qlm的均值 这样考虑了更多的环境情况 
def calculate_Qlm(center_pos, neibor_pos, l, m):
    center_pos,neibor_pos = np.array(center_pos), np.array(neibor_pos)
    nb = len(neibor_pos)
    sum_value = 0.0
    for i_atom in neibor_pos:
        sum_value += calculate_qlm()
    return np.sqrt(np.mean(np.abs(qlm_values)**2))
```

```python
# 计算 Ql(i)
def calculate_Ql(center_pos, neibor_pos, l):
    sum_Qlm = 0.0
    for m in range(-l,l+1):
        Qlm_values  = calculate_qlm(center_pos, neibor_pos, l, m)
        sum_Qlm += np.abs(Qlm_values)**2
    return np.sqrt(sum_Qlm * 4 * np.pi/(2*l+1))
```

```python
def read_xyz_frames(file_path):
    frames = []
    current_frame = None

    with open(file_path, 'r') as file:
        lines = file.readlines()
        line_index = 0

        while line_index < len(lines):
            line = lines[line_index].strip()

            if line.isdigit():
                # 读取原子数
                if current_frame:
                    frames.append(current_frame)
                num_atoms = int(line)
                current_frame = {"num_atoms": num_atoms, "atoms": []}
                line_index += 2  # 跳过标题行 "Atoms"
            else:
                # 读取原子坐标
                if current_frame:
                    atom_data = line.split()
                    if len(atom_data) == 4:
                        element, x, y, z = atom_data
                        current_frame["atoms"].append({
                            "element": element,
                            "x": float(x),
                            "y": float(y),
                            "z": float(z)
                        })
                line_index += 1

        if current_frame:
            frames.append(current_frame)

    return frames

# 使用之前的frames列表
frames = read_xyz_frames(file_path)

# 找出最大的x、y和z值
max_x = max([max(frame["atoms"], key=lambda atom: atom["x"])["x"] for frame in frames])
max_y = max([max(frame["atoms"], key=lambda atom: atom["y"])["y"] for frame in frames])
max_z = max([max(frame["atoms"], key=lambda atom: atom["z"])["z"] for frame in frames])

print("Maximum X value:", max_x)
print("Maximum Y value:", max_y)
print("Maximum Z value:", max_z)
```

```python
import numpy as np
from scipy.spatial.distance import cdist

def distance_with_pbc(coord1, coord2, box_size):
    delta = coord2 - coord1
    delta = delta - np.round(delta / box_size) * box_size
    dis  = np.linalg.norm(delta, axis=-1)
    if dis<1e-5:
        dis=100
    return dis

def find_neighbors(atom_list, center_elements, neighbor_elements, dis, box_size):
    center_indices = [i for i, atom in enumerate(atom_list) if atom[0] in center_elements]
    neighbor_indices = [i for i, atom in enumerate(atom_list) if atom[0] in neighbor_elements]

    neighbor_dict = {}

    for center_index in center_indices:
        center_coord = np.array(atom_list[center_index][1:])
        neighbors = []
        for j in neighbor_indices:
            neighbor_coord = np.array(atom_list[j][1:])
            dist = distance_with_pbc(center_coord, neighbor_coord, box_size)
            if np.min(dist) < dis:
                neighbors.append(atom_list[j])
        neighbor_dict[center_index] = {'center': atom_list[center_index], 'neighbors': neighbors}

    return neighbor_dict

```

# 使用KDT的dis计算近邻

```python
import numpy as np
from scipy.spatial import cKDTree

def kdt_dis(atom_data, box_size, center_types, neighbor_types, dis):
    """
    根据元素类型、坐标、盒子大小和距离阈值找出每个中心原子的近邻。

    参数：
    - atom_data: 包含原子类型、坐标的数组，每一行格式为 [type, x, y, z]
    - box_size: 盒子的大小，格式为 [box_x, box_y, box_z]
    - center_types: 中心原子的元素类型列表
    - neighbor_types: 近邻原子的元素类型列表
    - dis: 距离阈值

    返回：
    - neighbors_dict: 包含每个中心原子近邻的字典，格式为 {center_index: [neighbor1_index, neighbor2_index, ...]}
    """
    # 提取坐标和类型信息
    coordinates = atom_data[:, 1:]
    types = atom_data[:, 0]

    # 构建 KD 树
    kdtree = cKDTree(coordinates, boxsize=box_size)

    # 存储近邻的字典
    neighbors_dict = {}

    # 遍历每个原子
    for i in range(len(atom_data)):
        # 仅考虑中心原子类型
        if types[i] in center_types:
            # 查询距离小于阈值的近邻
            neighbor_indices = kdtree.query_ball_point(coordinates[i], r=dis)

            # 过滤近邻原子类型
            filtered_neighbors = [idx for idx in neighbor_indices if types[idx] in neighbor_types]

            # 去除中心原子自身
            filtered_neighbors = [idx for idx in filtered_neighbors if idx != i]

            # 存储结果
            neighbors_dict[i] = filtered_neighbors

    return neighbors_dict
```

# 使用KDT找到最近的n个近邻

import numpy as np
from scipy.spatial import cKDTree

def find_neighbors_cKDT(atom_data, box_size, center_types, neighbor_types, n_neighbors):
    """
    根据元素类型、坐标、盒子大小和近邻数量找出每个中心原子的近邻。

    参数：
    - atom_data: 包含原子类型、坐标的数组，每一行格式为 [type, x, y, z]
    - box_size: 盒子的大小，格式为 [box_x, box_y, box_z]
    - center_types: 中心原子的元素类型列表
    - neighbor_types: 近邻原子的元素类型列表
    - n_neighbors: 每个中心原子最近邻的数量

    返回：
    - neighbors_dict: 包含每个中心原子近邻索引的字典，格式为 {center_index: [neighbor_index1, neighbor_index2, ...]}
    """
    # 提取坐标和类型信息
    coordinates = atom_data[:, 1:]
    types = atom_data[:, 0]

    # 构建 KD 树
    kdtree = cKDTree(coordinates, boxsize=box_size)

    # 存储近邻的字典
    neighbors_dict = {}

    # 遍历每个原子
    for i in range(len(atom_data)):
        # 仅考虑中心原子类型
        if types[i] in center_types:
            # 查询最近的 n 个近邻
            _, neighbor_indices = kdtree.query(coordinates[i], k=n_neighbors)

            # 过滤近邻原子类型并存储索引
            filtered_neighbors = [idx for idx in neighbor_indices if types[idx] in neighbor_types]

            # 存储结果
            neighbors_dict[i] = filtered_neighbors

    return neighbors_dict

```python
import numpy as np
import matplotlib.pyplot as plt

# 示例数据，格式为 [x, y, z, q4, q6]
xyz_q4_q6 = si_xyz_q4_q6

# 提取 x、q4 和 q6 列
x_values = np.array([item[0] for item in xyz_q4_q6])
q4_values = np.array([item[3] for item in xyz_q4_q6])
q6_values = np.array([item[4] for item in xyz_q4_q6])

bins = 1.5

x_min_bin =  min(x_values)+4
x_max_bin =  max(x_values)-4
# 自定义 bin 边界数组
custom_bins = np.arange(x_min_bin, x_max_bin+ bins, bins)

# 使用 np.histogram 计算直方图
hist, bin_edges = np.histogram(x_values, bins=custom_bins)

# 使用 np.digitize 将 x 映射到 bins 中的索引
x_bin_indices = np.digitize(x_values, bin_edges)

# 初始化存储均值的数组
q4_mean_values = np.zeros(len(custom_bins) - 1)
q6_mean_values = np.zeros(len(custom_bins) - 1)

# 计算每个 bin 中 q4 和 q6 的均值
for i in range(1, len(custom_bins)):
    q4_mean_values[i - 1] = np.mean(q4_values[x_bin_indices == i])
    q6_mean_values[i - 1] = np.mean(q6_values[x_bin_indices == i])

# 创建图表
#fig, ax1 = plt.subplots()
# 设置图的大小
fig, ax1 = plt.subplots(figsize=(8, 6))

# 绘制均值随 x 的一维变化，改为折线图
ax1.set_xlabel('X')
ax1.set_ylabel('Mean Q4', color='tab:red')
ax1.plot(bin_edges[:-1], q4_mean_values, color='tab:red',marker='o',  linestyle='-', linewidth=2,markersize=4)
ax1.tick_params(axis='y', labelcolor='tab:red')

# 创建另一个坐标轴用于绘制 Mean Q6
ax2 = ax1.twinx()
ax2.set_ylabel('Mean Q6', color='tab:blue')
ax2.plot(bin_edges[:-1], q6_mean_values, color='tab:blue',marker='o', linestyle='-', linewidth=2,markersize=4)
ax2.tick_params(axis='y', labelcolor='tab:blue')

fig.tight_layout()
plt.show()
```
