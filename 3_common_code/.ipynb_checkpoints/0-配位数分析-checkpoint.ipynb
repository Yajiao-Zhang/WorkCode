```python
import numpy as np
import matplotlib.pyplot as plt
import copy
import time
```

```python
# æ–‡ä»¶å
# file_name = 'static_0k_out.xyz'
# æ–‡ä»¶ä½ç½®
file_path = r'D:\MD_file\Ge-Si\move_f\glass_300k_out.xyz'
#éœ€è¦è®¡ç®—é…ä½çš„å…ƒç´ åç§° å­—å…¸æ˜¯ç”¨æ¥è®¡æ•°è¿™ä¸ªå…ƒç´ åœ¨Aé™„è¿‘å‡ºç°çš„æ¬¡æ•° 
all_element_dic = {'O':0,'F':0,'Si':0}

#éœ€è¦åˆ†æçš„å…ƒç´ ä»¥åŠç¬¬ä¸€è¿‘é‚»çš„è·ç¦»
A_dis = 2.5
name_A = 'Si'

# åˆ†æçš„èµ·å§‹å¸§ã€ç»ˆæ­¢å¸§ã€ä»¥åŠæ­¥é•¿
start_frame = 0
steps_frame = 1
final_frame = 1

# æ˜¯å¦å°†åˆ†æç»“æœå­˜å…¥æ–‡ä»¶
save_result = 0
save_path = r'C:\Users\zhang\Desktop\50sap_out\\'
#save_name = file_name[:5] + '_ana'
#save_name+='.txt'
# æ˜¯å¦åˆ›å»ºæ–°å¸§  è½¬æ¢æ ¼å¼
creat_new_frame = 0
out_type = 'xyz'
```

```python
# æ‰“å¼€æ–‡ä»¶å¹¶è¯»å–æ‰€æœ‰è¡Œæ•°æ®
with open(file_path, 'r') as file:
    lines = file.readlines()
int(lines[0])
# æ¯ä¸€é¡µçš„åŸå­æ•°
num_atom =int(lines[0])
# å®šä¹‰ä¸€é¡µçš„è¡Œæ•°
lines_per_page = num_atom + 2
# è®¡ç®—æ€»é¡µæ•°
total_pages = (len(lines) + lines_per_page - 1) // lines_per_page

if final_frame>total_pages:
    print('Input Error!')
```

```python
def cal_ppp_distance(pos1, pos2, box_size):
    # è®¡ç®—ä¸¤ä¸ªä½ç½®ä¹‹é—´çš„æœ€å°è·ç¦»ï¼Œè€ƒè™‘å‘¨æœŸæ€§è¾¹ç•Œæ¡ä»¶
    pos1,pos2,box_size = np.array(pos1),np.array(pos2),np.array(box_size)
    # è®¡ç®—ä½ç½®å·®çŸ¢é‡
    delta = pos2 - pos1
    
    # åº”ç”¨æœ€å°å›¾åƒçº¦æŸï¼Œç¡®ä¿è·ç¦»åœ¨å‘¨æœŸå†…æœ€çŸ­
    delta = delta - np.round(delta / box_size) * box_size
    # è®¡ç®—è·ç¦»
    distance = np.linalg.norm(delta)
    # å¦‚æœæ˜¯åŸå­æœ¬èº«
    if distance<1e-5:
        distance=float('inf')
    return distance
```

```python
def draw_flower():
    pattern = [
        " ğŸŒ¹ğŸŒ¹ğŸŒ¹  ğŸŒ¹   ğŸŒ¹    ğŸŒ¹ ",
        " ğŸŒ¹  ğŸŒ¹  ğŸŒ¹  ğŸŒ¹     ğŸŒ¹ ",
        " ğŸŒ¹  ğŸŒ¹  ğŸŒ¹ğŸŒ¹ğŸŒ¹     ğŸŒ¹ ",
        " ğŸŒ¹  ğŸŒ¹  ğŸŒ¹   ğŸŒ¹       ",
        " ğŸŒ¹ğŸŒ¹ğŸŒ¹  ğŸŒ¹    ğŸŒ¹   ğŸŒ¹  "
    ]

    for row in pattern:
        print(row)
```

```python
# # è®°å½•æ¯ä¸€ä¸ªå¸§æ•°å¤„ç†åå¾—åˆ°çš„ä¿¡æ¯ï¼Œç„¶ååŠ å’Œ
# å¾—åˆ°å…ƒç´ Açš„æ•°é‡
num_A_atom = 0
# è®°å½•Aæ€»çš„è¿‘é‚»å’ŒCnæ•°
sum_A_neibor = 0
sum_A_Cn = np.zeros(100)
# å­—å…¸æ˜¯ç”¨æ¥è®¡æ•°çš„
sum_element_dic = copy.deepcopy(all_element_dic)
# ç›’å­å°ºå¯¸
box_size = []
# å¾ªç¯æ¬¡æ•°
count_cycle = 0 
```

```python
# å¼€å§‹è®¡æ—¶
start_time = time.time()
# 
max_x,max_y,max_z = 0,0,0
################### å¼€å§‹å¾ªç¯å¤„ç†æ¯ä¸€å¸§ #################
for frame in range(start_frame,final_frame,steps_frame):
    
    count_cycle+=1
    
    data = lines[frame*lines_per_page:(frame+1)*lines_per_page]
    ##  å»é™¤è¡¨å¤´
    data = data[2:]
    ##  å…ˆå¯¹æ•°æ®å¤„ç†ä¸€é
    new_data = []
    ana_element_list = []
#     # å…ˆæŠŠæ•°æ®åˆ‡ç‰‡ï¼Œå†åˆ†æ
#     data = data.split()
#     for i in range(0,len(data),4):
#         temp_x,temp_y,temp_z =data[i+1],data[i+2],data[i+3]
#         max_x,max_y,max_z = max(temp_x,max_x),max(temp_y,max_y),max(temp_z,max_z)
#         new_data.append([atom[0],temp_x,temp_y,temp_z])
    for atom in data:
        # è½¬æ¢æ ¼å¼
        atom = atom.split()
        temp_x,temp_y,temp_z = float(atom[1]),float(atom[2]),float(atom[3])
        max_x,max_y,max_z = max(temp_x,max_x),max(temp_y,max_y),max(temp_z,max_z)
        new_data.append([atom[0],temp_x,temp_y,temp_z])
        # å¦‚æœæ˜¯è¦åˆ†æçš„å…ƒç´ åˆ™è®°å½•ä¸‹æ¥
        if atom[0] == name_A:
            ana_element_list.append([atom[0],float(atom[1]),float(atom[2]),float(atom[3])])
    ## å…¨å±€å˜é‡
    box_size = [max_x,250,max_z]
    print(f'æ­¤å¤„æ›´æ”¹äº†ä¸‹boxsizeï¼Œ{box_size}')
    num_A_atom = len(ana_element_list)
    ## é…ä½æ•°è®°å½•çš„æ•°æ®è¡¨
    # è®°å½•è¿™ä¸ªAæ˜¯å‡ é…ä½
    temp_cn_list = np.zeros(100)
    # æ¯ä¸€æ¬¡è¦æ‹·è´ä¸€æ¬¡å­—å…¸ ä¾¿äºé‡æ–°è®¡æ•°
    frame_element_dic = copy.deepcopy(all_element_dic)
    ## å¼€å§‹åˆ†æ
    # å¯¹æ‰€æœ‰AåŸå­è¿›è¡Œåˆ†æ
    for ana_atom in ana_element_list:
        # è¿™æ˜¯åˆ†ææ¯ä¸€ä¸ªAåŸå­
        temp_cn = 0
        for neighber_atom in new_data:
            # åªå…³æ³¨äºdicä¸­æœ‰æ²¡æœ‰è¿™ä¸ªå…ƒç´ 
            if (neighber_atom[0] not in all_element_dic):
                continue
            atom1 = [ana_atom[1],ana_atom[2],ana_atom[3]]
            atom2 = [neighber_atom[1],neighber_atom[2],neighber_atom[3]]
            if cal_ppp_distance(atom1,atom2,box_size)<A_dis:
                temp_cn+=1
                # å†è®°å½•è¿™ä¸ªè¿‘é‚»çš„ç§ç±» å…ˆå–å‡ºåŸæœ‰å€¼ï¼Œå†åŠ ä¸€
                temp = frame_element_dic[neighber_atom[0]]
                frame_element_dic[neighber_atom[0]] = temp+1
        # æœ¬å¾ªç¯ç»“æŸ è®°å½•ä¸‹è¿™ä¸ªAå…ƒç´ çš„Cnå€¼   
        temp_cn_list[temp_cn]+= 1
    #####################å·²ç»å¾—åˆ°äº†è¿™ä¸€å¸§çš„æ‰€æœ‰æ•°æ®########################   
    # ç¬¬ä¸€é¡¹ Cnç´¯åŠ 
    for i in range(100):
        sum_A_Cn[i] += temp_cn_list[i]
    # ç¬¬äºŒé¡¹ æ€»çš„é…ä½æ•°ç´¯åŠ 
    for i in frame_element_dic:
        # å¦‚æœè¡¨æ˜ä¸è®¡æ•° åˆ™è·³è¿‡
        if frame_element_dic[i] == -1:
            continue
        sum_element_dic[i] += frame_element_dic[i]

# ç»“æŸè®¡æ—¶
end_time = time.time()
# è®¡ç®—è¿è¡Œæ—¶é—´
run_time = end_time - start_time

# æ‰“å°è¿è¡Œæ—¶é—´
print("ç¨‹åºè¿è¡Œæ—¶é—´ï¼š", run_time, "ç§’")
```

```python
########  è®¡ç®—å®Œæˆ è¾“å‡ºç»“æœ  ##############
print('file_name: ' + file_path.split('\\')[-1])
print('æˆ‘ä»¬å…³æ³¨çš„æ˜¯ '+ name_A +' çš„é…ä½æ•°ï¼Œé‡‡å–çš„æˆªæ–­è·ç¦»ä¸º '+str(A_dis))
print(name_A+'æœ‰'+ str(num_A_atom) +'ä¸ªï¼Œå®ƒçš„Cnåˆ†å¸ƒå¦‚ä¸‹ï¼š')
for i in range(100):
    if sum_A_Cn[i] != 0 :
        temp_x = sum_A_Cn[i]//count_cycle
        pro = temp_x/num_A_atom*100
        print(f'C' +str(i)+': '+str(sum_A_Cn[i]//count_cycle).center(8)+ '    å æ¯”ï¼š '+str(round(pro,2))+'%')
# å†æ¥åˆ†æAçš„å¹³å‡é…ä½æ•°
sum_cn = 0
for i in sum_element_dic:
    if sum_element_dic[i] != -1:
        sum_cn += sum_element_dic[i]
print(name_A+'çš„å¹³å‡Cnæ˜¯ï¼š '+str(round(sum_cn/(num_A_atom*count_cycle),2)))

cn = round(sum_cn/(num_A_atom*count_cycle),3)

# åˆ†æå„ä¸ªå…ƒç´ çš„é…ä½å æ¯”
print('å…ƒç´ åˆ—è¡¨ä¸­çš„æ€»ä½“é…ä½æƒ…å†µå¦‚ä¸‹ï¼š')
for i in sum_element_dic:
    if sum_element_dic[i]== -1:
        pass
        #print(i + " å…ƒç´ ä¸è®¡å…¥é…ä½è®¡ç®—ï¼")
    else:
        print(i + ' : '+'é…ä½æ•°é‡ï¼š'+str(sum_element_dic[i]//count_cycle)+' å æ¯”ï¼š '+ str(round(sum_element_dic[i]/sum_cn*100,1))+'%',' å‡å€¼:',str(round(cn*(sum_element_dic[i]/sum_cn),3)))
```
