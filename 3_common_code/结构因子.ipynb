```python
import numpy as np
from scipy.integrate import simpson  # 使用辛普森法则进行数值积分

def Sij(Q, gij_r, rho_j, r_values):
    """
    计算Sij(Q)值，根据公式

    Args:
        Q (float): 散射矢量
        gij_r (array): 配对分布函数 g_ij(r) 数值数据
        rho_j (float): 原子类型 j 的密度
        r_values (array): r 的离散值

    Returns:
        float: 计算得到的 Sij(Q) 值
    """
    # 计算 4 * pi * r^2 * (gij(r) - 1) * sin(Q * r) / (Q * r)
    integrand = 4 * np.pi * r_values**2 * (gij_r - 1) * np.sin(Q * r_values) / (Q * r_values)

    # 使用辛普森法则进行数值积分
    integral_value,err = simpson(integrand, r_values)

    # 计算 Sij(Q)
    Sij_value = 1 + rho_j * integral_value

    return Sij_value

# 示例数据
Q = 1.0  # 假设的散射矢量
rho_j = 0.5  # 假设的密度值

# 假设的配对分布函数gij(r)，这里是示例数据
r_values = np.linspace(0.001, 10, 1000)  # 从 0.001 到 10 的离散 r
gij_r = 1 + 0.1 * np.sin(r_values)  # 这里gij(r) - 1 只是一个示例函数

# 计算 Sij(Q)
Sij_value = Sij(Q, gij_r, rho_j, r_values)
print(f"Sij(Q) = {Sij_value}")
```

```python
def pair_distribution_function(elements, coordinates, box_size, r_max=10, dr=0.1):
    """计算配对分布函数 g_ij(r)"""
    n_atoms = len(elements)
    n_bins = int(r_max / dr)
    r_bins = np.linspace(dr, r_max, n_bins)
    g = np.zeros(n_bins)
    rho = n_atoms / np.prod(box_size)  # 原子数密度
    
    # 计算原子对间距离
    for i in range(n_atoms):
        for j in range(i+1, n_atoms):  # 只计算对称对
            dist = np.linalg.norm(coordinates[i] - coordinates[j])
            if dist < r_max:
                bin_idx = int(dist / dr)
                g[bin_idx] += 2  # 每对原子贡献2（i-j和j-i）

    # 归一化并计算 g(r)
    g /= (rho * 4 * np.pi * r_bins**2 * dr)
    return r_bins, g
```

```python
def structure_factor(Q, r_bins, g, box_size, r_max=10):
    """计算中子结构因子"""
    S = np.zeros(len(Q))
    R = box_size[0] / 2  # 最大积分距离设置为盒子尺寸的一半
    for i, q in enumerate(Q):
        integrand = lambda r: (4 * np.pi * r**2 * (g[np.digitize(r, r_bins)] - 1) *
                               np.sin(q * r) / (q * r))
        # 使用数值积分计算
        result, _ = quad(integrand, 0, r_max)
        S[i] = 1 + result
    return S
```

```python

def calculate_neutron_structure_factor(elements, coordinates, box_size, Q_vals):
    """计算整个系统的中子结构因子"""
    n_atoms = len(elements)
    S_total = np.zeros(len(Q_vals))
    
    # 计算配对分布函数
    r_bins, g = pair_distribution_function(elements, coordinates, box_size)
    
    # 计算每对原子的贡献
    for i in range(n_atoms):
        for j in range(i+1, n_atoms):
            element_i = elements[i]
            element_j = elements[j]
            b_i = b[element_i]  # 中子散射长度
            b_j = b[element_j]
            c_i = c[element_i]  # 权重
            c_j = c[element_j]
            
            # 计算部分结构因子
            S_ij = structure_factor(Q_vals, r_bins, g, box_size)
            S_total += c_i * b_i * c_j * b_j * S_ij
    
    return S_total
```

```python
# 示例用法
xyz_file = 'your_xyz_file.xyz'  # 替换为你的XYZ文件路径
Q_vals = np.linspace(0.1, 10, 100)  # Q值的范围

elements, coordinates, box_size = read_xyz(xyz_file)
S = calculate_neutron_structure_factor(elements, coordinates, box_size, Q_vals)

# 绘制结果
plt.plot(Q_vals, S)
plt.xlabel('Q (Å⁻¹)')
plt.ylabel('S(Q)')
plt.title('Neutron Structure Factor')
plt.show()
```
