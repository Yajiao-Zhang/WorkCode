```python
import re

#输入信息
#有几种元素
element_num = 3
box_list = []
#元素名和它的id
element_dic = {'O':1,'Al':3,'P':4}
O_list, Na_list, Al_list, P_list,Si_list,Li_list,Nb_list,Zn_list,B_list = [], [], [], [], [], [], [], [],[]

path = 'C:\\Users\\zhang\\Desktop\\cif文件\\B4O7Zn1_ICSD_412688.cif'
file = open(path)
data = file.read()
file.close()
```

```python
#解析abc
box_list = []
box_list.clear()
box_list.extend(re.findall(r'_cell_length_[abc].*?(\d{1,2}.\d{0,9})',data))
box_list.extend(re.findall(r'_cell_angle.*?(\d{1,2}.\d{0,9})',data))

if float(box_list[3])==90.0 and float(box_list[4])==90.0 and float(box_list[5])==90.0:
    print('晶体为立方体，直接输出结果。')
else:
    print('晶体为斜方晶，需要后续处理！')
```

```python
#解析各元素

O_list.clear()
O_list.extend(re.findall('O.*?(-{0,1}0\.\d{1,15}).*?(-{0,1}0\.\d{1,15}).*?(-{0,1}0\.\d{1,15})', data, re.M))

B_list.clear()
B_list.extend(re.findall('B.*?(-{0,1}0\.\d{1,15}).*?(-{0,1}0\.\d{1,15}).*?(-{0,1}0\.\d{1,15})', data, re.M))

# Na_list.clear()
# Na_list.extend(re.findall('Na.*?(-{0,1}0\.\d{1,15}).*?(-{0,1}0\.\d{1,15}).*?(-{0,1}0\.\d{1,15})', data, re.M))

# Al_list.clear()
# Al_list.extend(re.findall('Al.*?(-{0,1}0\.\d{1,15}).*?(-{0,1}0\.\d{1,15}).*?(-{0,1}0\.\d{1,15})', data, re.M))
# P_list.clear()
# P_list.extend(re.findall('P.*?(-{0,1}0\.\d{1,15}).*?(-{0,1}0\.\d{1,15}).*?(-{0,1}0\.\d{1,15})', data, re.M))
# Si_list.clear()
# Si_list.extend(re.findall('Si.*?(-{0,1}0\.\d{1,15}).*?(-{0,1}0\.\d{1,15}).*?(-{0,1}0\.\d{1,15})', data, re.M))

# Li_list.clear()
# Li_list.extend(re.findall('Li.*?(-{0,1}0\.\d{1,15}).*?(-{0,1}0\.\d{1,15}).*?(-{0,1}0\.\d{1,15})', data, re.M))
# Nb_list.clear()
# Nb_list.extend(re.findall('Nb.*?(-{0,1}0\.\d{1,15}).*?(-{0,1}0\.\d{1,15}).*?(-{0,1}0\.\d{1,15})', data, re.M))
# Zn_list.clear()
Zn_list.extend(re.findall('Zn.*?(-{0,1}0\.\d{1,15}).*?(-{0,1}0\.\d{1,15}).*?(-{0,1}0\.\d{1,15})', data, re.M))
```

```python
#输出lattice
print('lattice          custom   1.0   &')
print('a1 {} 0.0 0.0 a2 0.0 {} 0.0 a3 0.0 0.0 {} &'.format(box_list[0], box_list[1], box_list[2]))
for i in O_list:
    print('basis  {}  {}  {}  &'.format(i[0], i[1], i[2]))
    
# for i in Na_list:
#     print('basis  {}  {}  {}  &'.format(i[0], i[1], i[2]))
# for i in Al_list:
#     print('basis  {}  {}  {}  &'.format(i[0], i[1], i[2]))
# for i in P_list:
#     print('basis  {}  {}  {}  &'.format(i[0], i[1], i[2]))
# for i in Si_list:
#     print('basis  {}  {}  {}  &'.format(i[0], i[1], i[2]))
for i in B_list:
    print('basis  {}  {}  {}  &'.format(i[0], i[1], i[2]))
for i in Zn_list:
    print('basis  {}  {}  {}  &'.format(i[0], i[1], i[2]))   
 
# for i in Li_list:
#     print('basis  {}  {}  {}  &'.format(i[0], i[1], i[2]))
# for i in Nb_list:
#     print('basis  {}  {}  {}  &'.format(i[0], i[1], i[2]))

```
