```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,100)
```

```python
y = 5*x+2
```

```python
noise = np.random.rand(100)*0.8
```

```python
y1 = y+noise
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y1,c='green')
plt.xticks(range(-5,6))
plt.yticks(range(-25,35,10))
plt.xlabel(' X ')
plt.ylabel(' F(x) ')
plt.title("Linear function")
```

```python
file_name = 'line_data.txt'
# 使用with语句写入文件
with open(file_name, "w") as file:
    for i in range(100):
        file.write(str(x[i])+' '+ str(y[i])+' \n')
```
