```python
# 先找出所有的Si_list，在第一帧里计算出它的近邻的十个index
# Si
center_element = '4'
# O
neighbor_element = '1'

# 近邻的数量
neighbor_num = 30
```

```python
# 再把排好序的原子进行计算 这样直接用列表中的index就可以了
max_x,max_y,max_z = 0,0,0
for i,i_atom in enumerate(frames[0]):
    max_x,max_y,max_z = max(max_x,float(i_atom.x)),max(max_y,float(i_atom.y)),max(max_z,float(i_atom.z))
box_size = [max_x+0.01,max_y+0.01,max_z+0.01]
print(box_size)
```

```python
# 找到了最近的十个nei 然后在之后的每一帧当中计算dis

start_frame = 0
end_frame = 1
step_frame = 1

All_result = []

# 直接计算每个帧数中Si最近的十个O周围的原子吧 然后再统计
for frame_i in range(start_frame,end_frame,step_frame):
    # 转换格式 得出每一帧的数据
    center_list = []
    neighbor_list = []
    # 加上list_index 和atom_index
#     count_si,count_o=0,0
    for i,i_atom in enumerate(frames[frame_i]):
        
        if i_atom.atom_type == center_element:
            # list_index,atom_index, xyz
            center_list.append([0, i_atom.atom_id,float(i_atom.x),float(i_atom.y),float(i_atom.z)])
            
        elif i_atom.atom_type == neighbor_element:
            neighbor_list.append([0, i_atom.atom_id,float(i_atom.x),float(i_atom.y),float(i_atom.z)])
            
    # 甚至是不用排序 只要一对一对应就行
    #sorted_list = sorted(two_dimensional_list, key=lambda x: x[1])
    #print(center_list[0])
    # 按照每帧内的原子索引从小到大排序
    center_list = sorted(center_list, key=lambda x: x[1])
    neighbor_list = sorted(neighbor_list, key=lambda x: x[1])
    # 再加上list_index  
    for i in range(len(center_list)):
        center_list[i][0] = i
    for j in range(len(neighbor_list)):
        neighbor_list[j][0] = j
    
    # 提取出数据
    Si_list = center_list
    O_list = neighbor_list
    box = box_size
    # 开始计算
    # 从Si_list和O_list中提取坐标数据
    Si_coords = np.array([coord[2:] for coord in Si_list])
    O_coords = np.array([coord[2:] for coord in O_list])

    # 创建一个KD树以便快速查找最近的氧原子，并考虑PBC
    o_kdtree = cKDTree(O_coords, boxsize=box)

    # 初始化一个空的列表来存储结果
    si_nearest_o = []

    # 遍历每个硅原子
    for si_index, si_coord in enumerate(Si_coords):
        # 查询最近的十个氧原子
        distances, indices = o_kdtree.query(si_coord, k=neighbor_num)

        # 添加结果到si_nearest_o列表，包括Si的索引
        # 输出结果 list_index  type
        si_nearest_o.append((center_list[si_index][0],center_list[si_index][1], si_coord, [[O_list[i][0],O_list[i][1]] for i in indices], distances))
    # 最后总结数据
    All_result.append(si_nearest_o)
```

```python
mean_list = []
std_list = []
# 然后计算每一行的均值和标准差
for i in all_si_o:
    # 每一个si
    temp_mean = []
    temp_std = []
    for j in i:
        # 计算每行的均值
        row_means = np.mean(j)
        temp_mean.append(row_means)
        # 计算每行的标准差
        row_stddevs = np.std(j)
        temp_std.append(row_stddevs)
    mean_list.append(temp_mean)
    std_list.append(temp_std)
```
