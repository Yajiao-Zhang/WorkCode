```python
# define Potential function
# 然后就是势函数的作用类型 
# O-O就是0,O-Si为1，Si-Si为2
def Teter(r_ij,r_type,Z_i,Z_j):
    ## 常见参数
    epsilon_0, e_spot, pi = 8.854187817*(10**-12),1.602176634*(10**-19),np.pi
    ## 第一步先给出参数
    A_ij,row_ij,C_ij,B_ij,D_ij,n_ij,r0_ij = None,None,None,None,None,None,None
    # 如果是O-O
    if r_type=='O-O':
        A_ij,row_ij,C_ij,B_ij,D_ij,n_ij,r0_ij = 2029.2204,0.343645,192.58,45.868,-0.333,3.402,1.903
    # 如果是O-Si
    elif r_type=='O-Si':
        A_ij,row_ij,C_ij,B_ij,D_ij,n_ij,r0_ij = 13702.9050,0.193817,54.681,28.950,-3.059,3.932,1.168
    # 如果是Al-O:
    elif r_type=='O-Al':
        A_ij,row_ij,C_ij,B_ij,D_ij,n_ij,r0_ij = 12201.4170,0.195628,31.997,51.728,-10.197,3.179,1.024
    ## 第二部，开始计算作用
    elif r_type=='O-Na':
        A_ij,row_ij,C_ij,B_ij,D_ij,n_ij,r0_ij =4383.7555,0.243838,30.700,48.264,-4.730,2.893,1.172
    elif r_type=='O-P':
        A_ij,row_ij,C_ij,B_ij,D_ij,n_ij,r0_ij =26655.4720,0.181968,86.856,28.566,-3.408,4.635,1.168
    elif r_type=='O-Zr':
        A_ij,row_ij,C_ij,B_ij,D_ij,n_ij,r0_ij =17943.3840,0.226627,127.650,108.655,-8.481,3.305,1.224
    else:
        print('{} Bond not in this sets'.format(r_type))    
        return 0
    V_ij = None
    # 如果是在正常的作用范围内
    if r_ij>r0_ij:
        V_ij = (Z_i*Z_j*e_spot**2/(4*pi*epsilon_0*r_ij)) +A_ij*np.exp(-r_ij/row_ij)-C_ij/r_ij**6
    # 如果不是在这个范围内
    else:
        V_ij = B_ij/(r_ij**n_ij)+D_ij*r_ij**2
    return V_ij
```

```python
# 定义距离的计算
def cal_3d_distance(pos1,pos2,box_size):
    pos1,pos2,box_size = np.array(pos1),np.array(pos2),np.array(box_size)
     # 计算每个坐标的差值
    delta = pos2-pos1
    # 应用周期性边界条件
    delta = delta - np.round(delta / box_size) * box_size
    # 计算欧几里得距离
    distance = np.linalg.norm(delta)
    # 如果距离等于0了 就表示是本身
    if distance<1e-4:
        distance = 0
    return distance
# 定义距离的计算
def cal_2d_distance(x1,x2,len_box):
     # 计算每个坐标的差值
    delta = x2-x1
    # 应用周期性边界条件
    delta = delta - np.round(delta / len_box) * len_box
    # 如果距离等于0了 就表示是本身
    if abs(delta)<1e-4:
         delta = 0
    return abs(delta)
```
