```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# def parameters
PARA_Kb = 8.617333262145*10**(-5)  # eV·K−1
```

```python
def delta_fu(T,Tm,delta_hm):
    return delta_hm*(Tm-T)/Tm
def thermodynamic(T,Tm,delta_hm):
    delta_u = delta_fu(T,Tm,delta_hm)
    return 1-np.exp(-delta_u/PARA_Kb/T)
```

```python
def kinetic_BGJ(T):
    return (3*PARA_Kb*T)**0.5
```

```python
x_origin = np.linspace(300,3000,1000)
```

```python
y = kinetic_BGJ(x_origin)
plt.title('K(t) vs. BGJ_model')
plt.plot(x_origin,y)
```

```python
y = kinetic_BGJ(x_origin)*thermodynamic(x_origin,2100,0.062)
plt.title('V(T) vs. BGJ_model')
plt.plot(x_origin,y)
```

```python
y = thermodynamic(x_origin,2100,0.062)
# "axis vertical line" 和 "axis horizontal line" 
plt.plot(x_origin,y)
plt.title('thermodynamic')
plt.axvline(2100,color = 'r',label = 'Tm=2100 K')
plt.axvline(1000,color = 'black',label = 'Tm=2100 K')
plt.ylim(-0.2,1)
#plt.axhline(0.5,color = 'b',linestyle = '--',label = 'Y=0.1')
plt.axhline(0,color = 'b',label = 'Y=0')
plt.legend()
```
